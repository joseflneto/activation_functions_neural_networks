```python
import seaborn as sns
import numpy as np
```

```python
x = [i for i in range(-5, 6)]
```

# Step function

## Linearly separable problems
## Returns 0 or 1

```python
def step_function(x):
    
    if x >= 1:
        return 1
    return 0
```

```python
y = list(map(step_function, x))

sns.lineplot(x = x, y = y)
```

# Sigmoid function

## Binary classification problems
## Non Linearly separable problems
## Returns between 0 and 1

```python
def sigmoid(x):
    return 1 / (1 + (np.exp(-x)))
```

```python
y = list(map(sigmoid, x))

sns.lineplot(x = x, y = y)
```

# Hyperbolic Tangent

## Returns between -1 and 1
## Binary problems
## Negative values of input will be mapped with negative outputs

```python
def hyperbolic_tangent(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
y = list(map(hyperbolic_tangent, x))

sns.lineplot(x = x, y = y)
```

# ReLu function

## Returns or 0 or a value greater than 0 (no max values)
## Widely used today in CNN and deep learning, tends to better results
## Does not perform well when there are many negative values on input data

```python
def reLu(x):
    return max(0, x)
```

```python
y = list(map(reLu, x))

sns.lineplot(x = x, y = y)
```

# Linear

## Widely used on regression problems

```python
def linear(x):
    return x
```

```python
y = list(map(linear, x))

sns.lineplot(x = x, y = y)
```

# Softmax

## One of the most important
## Multiclass problems
## Multiple inputs on the function, the function will returns the probability for each input (in pratice, the probability for each class)

```python
def softmax(x : list):
    x_exp = np.exp(x)
    return x_exp / x_exp.sum()
```

```python
y = softmax(x)
sns.barplot(x = x, y = y)
```
